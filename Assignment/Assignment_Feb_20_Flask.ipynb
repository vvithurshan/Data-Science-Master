## Q1. Explain GET and POST methods.

GET and POST methods are HTTP methods used in webapplications. 
GET method are used to retrive the data requested by the user. Here, the requested information is passed through the URL of the web page. The parameters are appended at the end of the URL after **?** symbol, and separated by **&**. Howerver, sending requests through the GET method is not secure as it can be seen in the URL.

POST methods are used to submit documents such as form, imgage, or file to a specific location. Unlike, GET methods, POST methods send encoded data in the body of the page. Therefore, it is not visible to anyone.

## Q2. Why is request used in Flask?

A request is an instance of the Request class used to access incoming HTTP requests. It provides consistent way to access data sent through HTTP requests, regardless of the methods ( GET, POST). 
- request.form attribute is used to access the form as a dictionary file.
- request.args attribute is used to access the URL parameters as a dictionary file.
- request.file attribute is used to access the file as a dictionary file.

## Q3. Why is redirect() used in Flask?

A redirect() function is used to redirect the user to a different URLs. It is used for the following occasions,
- When a user tries to access a protected page that require authentication, a user will be prompted to a login page.
- When a web page is moved to a different URL, redirect() function is used to redirect the incoming request to the new URL.
- When an error or exception occurrs in the application, redirect() function is used to redirect the page to a different page.

Templates are written in the language called template. It allows applications to generate dynamic applications based on the provided data.

```python
from flask import Flask, request, render_template, jsonify

app = Flask(__name__)

@app.route("/")
def home_page():
    return render_template("index.html")

## currency conversion data
currency_data = {
    'USD': {'USD': 1, 'EUR': 0.94, 'GBP': 0.83},
    'EUR': {'USD': 1.07, 'EUR': 1, 'GBR': 0.88},
    'GBP': {'USD': 1.21, 'EUR': 1.13, 'GBP': 1}
}


## API to convert currencies
@app.route("/postman_action", methods = ['POST'])
def convert2():
    if ( request.method == 'POST'):
        Amount = request.json['amount']
        From_currency = request.json['from_currency']
        To_currency = request.json['to_currency']
        Amount_converted = (currency_data[From_currency][To_currency]) * float(Amount)
        result = "{} {} = {} {}".format(From_currency, Amount, To_currency, Amount_converted)
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
