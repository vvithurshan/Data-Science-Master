## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in python. It is used to build small to medium sized web applications.

**Advantages**

1. Easy to learn and use - Its syntax are very easy to learn and use. Therefore, it is more flexible.

2. Lightweight and flexible - It does not need a lot of code to build the boilerplate code, making it very easy and flexible to use.

3. Built-in development server - It has built-in server. Therefore, applications can be tested easily.

4. Extensible - Flask supports many extensions and plugins. Therefore, more functionality can be added to the applications easily.

5. Template engine - It has a built-in template engine called Jinja2. Therefore, HTML pages with dynamic content can be developed easily.

6. RESTful request handling - It is well suited for building RESTful APIS.

## Q3. What is App routing in Flask? Why do we use app routes?

Routing refers to the process of mapping URLs to view functions that handle HTTP requests. 

Route is used to define the behabior of our application when a user access a specific URL. It allows us the map the incoming requests to a specific functions in our applications. Therefore, various types of requests from the user can be handled. 

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation <br>
Location: India <br>
Contact Detail: 999-999-9999

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route(("/welcome"))
def welcome():
    return "<H1> Welcome to ABC Corporation </H2>"

@app.route("/")
def ABC():
    return """<H2> Company Name: ABC Corporation </H2>
              <H2> Location: India </H2>
              <H2> Contact Detail: 999-999-9999 </H2>
                """

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

With the help of url_for() function, URL for different endpoint with proper parameters can be generated. Therefore, it allows user to access different parts of the application. 

In the example below, user_for() function has created URLs for root, home, and user function dynamically. 

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def root():
    return "Welcome"

@app.route("/home")
def home():
    return "Welcome to our service"

@app.route("/account/<username>")
def user(username):
    return "LOGIN %s" %username

with app.test_request_context():
    root_url = url_for('root')
    print('Root URL: ', root_url)

    home_url = url_for('home')
    print('HOME URL: ', home_url)

    user_url = url_for('user', username = "Vithurshan")
    print("User URL: ", user_url)
```
