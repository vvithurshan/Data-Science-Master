## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df = df.dropna()
```

```python
df['Price_numeric'] = df['Price'].str.replace("$","")
```

```python
df['Price_numeric'] = df['Price_numeric'].astype(float)
```

```python
df.drop('Price', inplace=True, axis=1)
```

```python
df.info()
```

```python
avg_price = df['Price_numeric'].groupby(df['Type']).mean()
```

```python
plt.bar(avg_price.index, avg_price.values)
```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs']=  df['Installs'].str.replace(",","").str.replace("+","").astype(int)
```

```python
total_installs = df['Installs'].groupby(df['App']).sum()
```

```python
total_installs = total_installs.sort_values(ascending=False)
```

```python
top_10_apps = total_installs.head(10)
```

```python
freq_table = pd.DataFrame({'App': top_10_apps.index, 'Total Installs': top_10_apps.values})
```

```python
print(freq_table)
```

```python
plt.bar(freq_table['App'], freq_table['Total Installs'])
plt.xticks(rotation=90)
plt.xlabel('App')
plt.ylabel('Total Installs')
plt.title('Top 10 Most Popular Apps by Number of Installs')
plt.show()
```
