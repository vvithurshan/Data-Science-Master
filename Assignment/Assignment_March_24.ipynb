## Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.


1. Fixed acidity, volatile acidity, and citric acid - The acidity level of wine is an imporant factor in deciding the quality of the wine. The acids affects the taste and pH which affects the color, oxidation stability, and consequentely the overall lifespan of the wine. There are two types of acids can be found in wine, fixed acids and volatile acids. Fixed acidity is a fundamental property of a wine, imparting the sourness and resistance to microbial activity. At the same time, volatile acidity indicates the spoilage of the wine. The level of citric acid indicates the freshness of the wine, though it promotes the growth of unwanted microbes. Therefore, the level of acidity is important for the quality of the wine. 

2. Residual sugar - residual sugar is one of the important componets that contributes to the sweetness of a wine in additon to the acidity and alcohol content of wine. Therefore, residual sugar is an important factor in deciding the quality of the wine.

3. chlorides - Chloride affects the taste and quality of the wine. It is also used as a preservative agent of wine. Additionally, it chelates metal ions and prevents browning reactions. However, excess level of chloride causes the growth of unwanted microbes.

4. Free and total sulfur dioxide - Sulfur dioxide is added to wine to prevent unwanted reactions. It is mainly present in two forms. The free sulfur dioxide acts as a preservative agent, while the bound form, which accounts for the most sulfur dioxide content, plays a less important role in wine. However, in certain conditions, bound form can release free sulfur sulform dioxide. 

5. Density - Density of the wine is a physical property that can affect the taste, alcohol content, viscosity, stability and clarity of the wine.

6. pH - pH level is a measure of the acidity of wine. It also affects the sensory properties, stability and quality of wine.

7. Sulphates - Sulphates prevent unwanted reactions, such as oxidation and browning. They also enhance the flavor and aroma of wine. However, some people may be allergic to sulphites and they reduce the beneficial compounds of wine.

8. alcohol - Alcohol is a key component of wine. It is responsible for various important properties of wine, such as stability and quality. It also affects the preservation, oxidation and lifespan of wine. In addition to that, it has both negative and positive effects on health. Therefore, the level of alcohol is important for the quality of wine.

## Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

In wine quality dataset, there was no missing data.

1. mean imputation 

    Advantages:

    - Easy to implement and computationally efficient
    - Preserves the sample size and avoids loss of data
    - Can be used for both numerical and categorical data
    
    Disadvantages:

    - May lead to biased estimates of the true values, especially if the missing values are not missing at random 
    - Reduces variability in the data and may underestimate the true standard deviation and covariance
    - May distort correlations between variables
    
2. Mode imputation
    Advantages:

    - Mode imputation is a simple technique that is easy to understand and implement.
    - Mode imputation can preserve the distribution of the data if the missing values are random and not dependent on the other variables.
    - Suitable for categorical data: Mode imputation is a suitable method for categorical data, where the values are limited to a small set of options.
    Disadvantages:

    - Mode imputation can introduce bias if the missing values are not randomly distributed, and the most frequent value is not representative of the entire dataset.
    - Mode imputation can decrease the variability in the data by replacing missing values with a single value.
    - Mode imputation ignores the correlation between variables, which can lead to inaccurate results if the missing values are dependent on other variables in the dataset.
    
3. Median imputation

    Advantages:

    - Median imputation is a simple and easy to understand method for dealing with missing values.
    - Median imputation is robust to the presence of outliers in the data, as the median is less sensitive to outliers than the mean.
    - Median imputation can be used for both numerical and categorical variables, unlike mean imputation which is only appropriate for numerical variables.
    Disadvantages:

    - Median imputation does not take into account any relationships between variables, so it may not be appropriate in situations where there are complex dependencies between variables.
    - Median imputation may result in biased estimates of the true population parameters, especially if the proportion of missing values is large.
    - Median imputation may result in an underestimation of the variability of the data, as it ignores any information contained in the missing values.

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("winequality-red.csv")
```

```python
df.columns[:-1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
DF = df[df.columns[:-1]]
pca = PCA()
pca.fit(DF)

explained_variance_ratio = pca.explained_variance_ratio_

plt.plot([i for i in range(len(explained_variance_ratio))], explained_variance_ratio, 'x')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Plot')
plt.show()
```

```python
import numpy as np
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.plot([i for i in range(len(explained_variance_ratio))], cumulative_variance_ratio, 'x')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Variance Plot')
plt.show()
```

```python
explained_variance_ratio
```

```python
cumulative_variance_ratio
```

### As shown above, first component explains 94.6 % of the variance in the data. Therefore, using only one component is more than enough to explain 90 % of the variance.
