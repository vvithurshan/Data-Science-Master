## Q1. How can you create a Bokeh plot using Python code?

Bokeh plot can be created by the following way

```python
import bokeh.io
import bokeh.plotting
!bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
```

```python
from bokeh.sampledata.iris import flowers
output_file("First.html")

p = figure(title = "First")
p.xaxis.axis_label  = "petal_length"
p.yaxis.axis_label = "petal_width"
p.scatter(x = flowers['petal_length'], y = flowers['petal_width'], color = 'blue', size=8)
show(p)
```

## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
import numpy as np
import random

x = np.random.randint(1, 10, 500)
y = np.random.randint(1, 10, 500)


color_book = ['red', 'green', 'blue', 'orange', 'yellow']

color = [color_book[random.randint(0,4)] for i in range(500)]


output_file("Glyphs.html")

p = figure()
p.title="Glyphs"
p.xaxis.axis_label = "X axis"
p.yaxis.axis_label = "Y axis"

data = dict(x = x, y = y, color = color)
Data = ColumnDataSource(data)

p.triangle('x', 'y', color = 'color', size = 8, source = Data)
show(p)
```

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Bokeh server is a way to create interactive plots that can be updated in real time. When the code below is executed, it will create a bokeh server application plot and update the data. It can be viewed by running Bokeh server and navigating to the URL with the application name.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
import random

data = dict(x = [], y = [])
Data = ColumnDataSource(data)

p = figure()
p.scatter(x='x', y='y', source=Data)
p.title("Bokeh server")

def update():
    new_data = dict(x=[random.random()*100], y=[random.random()*100])
    Data.stream(new_data, rollover=50)

curdoc().add_periodic_callback(update, 1000)

curdoc().add_root(p)
```

## Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

By executing the following code using Flask Bokeh plot can be embeded into a web page. 

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure(title="My Plot")
    p.line([1, 2, 3, 4, 5], [2, 4, 6, 8, 10])

    # Embed the plot components in the HTML template
    script, div = components(p)
    return render_template('index.html', plot_script=script, plot_div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
