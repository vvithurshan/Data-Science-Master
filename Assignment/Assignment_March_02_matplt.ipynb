## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plot is a type of plot that uses dots to represents values in graph.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
plt.title("Scatter Plot")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.scatter(x,y)
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() is a function in matplotlib used to create multiple plots within a single figure. Therefore, multiple plots can be arranged in a grid in a desired place by specifying the row and column. 

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0,0].plot(x1,y1)
axs[0,0].set_title('Line 1')
axs[0,1].plot(x2,y2)
axs[0,1].set_title('Line 2')
axs[1,0].plot(x3,y3)
axs[1,0].set_title('Line 3')
axs[1,1].plot(x4,y4)
axs[1,1].set_title('Line 4')
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plot is representation of categorical data with rectangular bars. In this graph, each bar represents a category and the frequency of each bar is represented with height.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(5,3))
plt.title("Profit of top companies")
plt.ylabel("Profit")
plt.xlabel("Companies")
plt.bar(company, profit)
```

### Horizontal

```python
plt.figure(figsize=(5,3))
plt.title("Profit of top companies")
plt.ylabel("Profit")
plt.xlabel("Companies")
plt.barh(company, profit)
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot represents the distribution of numerical data with quartiles. It consists of box and whiskers. A box shows the middle 50 % data, while the whiskers show the rest of the data except outpliers. In side the box, the median is marked by aline.

Box plots are commonly used in statistical analysis to compare the distribution of multiple data sets in a single data set.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()
```
