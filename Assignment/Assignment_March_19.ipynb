## Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

Principal Component Analysis (PCA) is a dimensionality reduction technique that enables us to identify correlations and patterns in a dataset so that it can be transformed into a dataset of significantly fewer dimensions without loss of any important information.

PCA works by identifying the principal components of the dataset, which are the directions in which the data varies the most. These principal components are then used to transform the data into a new coordinate system, where the first coordinate has the largest possible variance, the second coordinate has the second-largest variance, and so on.

For example, if we have a dataset with 100 features. We can use PCA to reduce the number of features to 10 while retaining most of the information in the original dataset. If we have a dataset of images with 1000 pixels. We can use PCA to reduce the number of pixels to 100 while retaining most of the information in the original images.

## Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform Feature Extraction using PCA. How many principal components would you choose to retain, and why?

The explained variance ratio of a principal component measures the proportion of the total variance in the data that is explained by that component.
For this purpose, random number are created with a hlep of numpy.

The plot shows the explained variance ratio of each principal component as a function of the component number. 

When deciding on the number of principal components to retain, one approach is to examine the cumulative sum of the explained variance ratios and choose the number of principal components that explain a sufficient amount of the total variance.

if we examine the cumulative variance plot, we can see that the first three principal components capture approximately 68 % of the total variance. Therefore, retaining the first three principal components may be a reasonable choice for this dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

n_samples = 1000

n_features = 5

X = np.random.rand(n_samples, n_features)

pca = PCA()
pca.fit(X)

explained_variance_ratio = pca.explained_variance_ratio_

plt.plot(np.array([1, 2, 3, 4, 5]), explained_variance_ratio, 'x')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Plot')
plt.show()
```

```python
import numpy as np
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.plot(np.array([1, 2, 3, 4, 5]), cumulative_variance_ratio, 'x')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Variance Plot')
plt.show()
```
